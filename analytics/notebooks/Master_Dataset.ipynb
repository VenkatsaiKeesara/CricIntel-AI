```python
import pandas as pd

# ==========================
# LOAD DATASETS
# ==========================

customers = pd.read_csv("../../data/raw/Customers.csv")
matches = pd.read_csv("../../data/raw/Matches.csv")
ticket_sales = pd.read_csv("../../data/raw/Ticket_Sales.csv")

teams = pd.read_csv("../../data/reference/teams.csv")
stadiums = pd.read_csv("../../data/reference/stadiums.csv")
seat_categories = pd.read_csv("../../data/reference/seat_categories.csv")
cities = pd.read_csv("../../data/reference/cities.csv")

print("✅ All datasets loaded successfully!")
```

```python
datasets = {
    "Customers": customers,
    "Matches": matches,
    "Ticket_Sales": ticket_sales,
    "Teams": teams,
    "Stadiums": stadiums,
    "Seat_Categories": seat_categories,
    "Cities": cities
}

for name, df in datasets.items():
    print("=" * 80)
    print(name)
    print("Shape:", df.shape)
    print("Columns:")
    print(df.columns.tolist())
    print()
```

```python
# ==========================================================
# CREATE CRICINTEL MASTER DATASET
# ==========================================================

# Step 1: Ticket Sales + Customers
master_df = ticket_sales.merge(
    customers,
    on="Customer_ID",
    how="left"
)

# Step 2: Add Match Details
master_df = master_df.merge(
    matches,
    on="Match_ID",
    how="left",
    suffixes=("", "_Match")
)

# Step 3: Add Seat Category Details
master_df = master_df.merge(
    seat_categories,
    on="Seat_Category",
    how="left"
)

# Step 4: Add State Information
master_df = master_df.merge(
    cities,
    on="City",
    how="left",
    suffixes=("", "_City")
)

print("✅ Master Dataset Created Successfully!")
print(f"Rows    : {master_df.shape[0]}")
print(f"Columns : {master_df.shape[1]}")

master_df.head()
```

```python
# ==========================================================
# MASTER DATASET VALIDATION
# ==========================================================

print("="*60)
print("MASTER DATASET SUMMARY")
print("="*60)

print(f"Rows      : {master_df.shape[0]}")
print(f"Columns   : {master_df.shape[1]}")

print("\nDuplicate Rows :", master_df.duplicated().sum())

print("\nMissing Values")
print(master_df.isnull().sum()[master_df.isnull().sum()>0])

print("\nData Types")
print(master_df.dtypes)
```
