```python
# =============================================================================
# MODULE 1 : LOAD MASTER DATASET
# Purpose:
# Load the cleaned analytical master dataset created from the previous notebook.
# This dataset will be used to engineer business features for analytics and AI.
# =============================================================================

import pandas as pd
import numpy as np

# Display Settings
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)

# Load Master Dataset
master_df = pd.read_csv("../../data/processed/cricintel_master_dataset.csv")

print("="*70)
print("MASTER DATASET LOADED SUCCESSFULLY")
print("="*70)
print(f"Rows    : {master_df.shape[0]}")
print(f"Columns : {master_df.shape[1]}")

master_df.head()
```

```python
# =============================================================================
# MODULE 2 : DATA TYPE CONVERSION
# Purpose:
# Convert date columns from object type to datetime format so that
# calendar-based business features can be created.
# =============================================================================

# Convert Date Columns
master_df["Booking_Date"] = pd.to_datetime(master_df["Booking_Date"])
master_df["Match_Date"] = pd.to_datetime(master_df["Match_Date"])
master_df["Date_of_Birth"] = pd.to_datetime(master_df["Date_of_Birth"])
master_df["Registration_Date"] = pd.to_datetime(master_df["Registration_Date"])

print("="*70)
print("DATE CONVERSION COMPLETED")
print("="*70)

print(master_df[
    ["Booking_Date",
     "Match_Date",
     "Date_of_Birth",
     "Registration_Date"]
].dtypes)
```

```python
# =============================================================================
# MODULE 3 : DATE & TIME FEATURES
# Purpose:
# Extract calendar-based business features for demand forecasting,
# booking trend analysis and executive reporting.
# =============================================================================

# -----------------------------
# 1. Booking Lead Time
# -----------------------------
master_df["Booking_Lead_Days"] = (
    master_df["Match_Date"] - master_df["Booking_Date"]
).dt.days

# -----------------------------
# 2. Booking Month
# -----------------------------
master_df["Booking_Month"] = master_df["Booking_Date"].dt.month_name()

# -----------------------------
# 3. Booking Weekday
# -----------------------------
master_df["Booking_Day_Name"] = master_df["Booking_Date"].dt.day_name()

# -----------------------------
# 4. Match Month
# -----------------------------
master_df["Match_Month"] = master_df["Match_Date"].dt.month_name()

# -----------------------------
# 5. Match Weekday
# -----------------------------
master_df["Match_Day_Name"] = master_df["Match_Date"].dt.day_name()

# -----------------------------
# 6. Match Quarter
# -----------------------------
master_df["Match_Quarter"] = "Q" + master_df["Match_Date"].dt.quarter.astype(str)

# -----------------------------
# 7. Weekend Match Flag
# -----------------------------
master_df["Weekend_Match"] = np.where(
    master_df["Match_Date"].dt.dayofweek >= 5,
    "Yes",
    "No"
)

print("="*70)
print("DATE & TIME FEATURES CREATED")
print("="*70)

print("New Features Added:")
print([
    "Booking_Lead_Days",
    "Booking_Month",
    "Booking_Day_Name",
    "Match_Month",
    "Match_Day_Name",
    "Match_Quarter",
    "Weekend_Match"
])

master_df.head()
```

```python
# =============================================================================
# MODULE 3 VALIDATION
# =============================================================================

new_features = [
    "Booking_Lead_Days",
    "Booking_Month",
    "Booking_Day_Name",
    "Match_Month",
    "Match_Day_Name",
    "Match_Quarter",
    "Weekend_Match"
]

master_df[new_features].describe(include="all")
```

```python
# =============================================================================
# MODULE 4 : BOOKING BEHAVIOR FEATURES
# Purpose:
# Create customer booking behaviour features to analyze
# booking patterns, booking size and booking value.
# =============================================================================

# ---------------------------------------------------
# Feature 1 : Booking Window
# ---------------------------------------------------
# Classify bookings based on how early they were made.

master_df["Booking_Window"] = np.select(
    [
        master_df["Booking_Lead_Days"] >= 30,
        master_df["Booking_Lead_Days"].between(15,29),
        master_df["Booking_Lead_Days"] < 15
    ],
    [
        "Early Booking",
        "Regular Booking",
        "Last Minute"
    ],
    default="Unknown"
)

# ---------------------------------------------------
# Feature 2 : Booking Size
# ---------------------------------------------------
# Categorize bookings by ticket quantity.

master_df["Booking_Size"] = np.select(
    [
        master_df["Quantity"] == 1,
        master_df["Quantity"] == 2,
        master_df["Quantity"] >= 3
    ],
    [
        "Single",
        "Couple",
        "Group"
    ],
    default="Unknown"
)

# ---------------------------------------------------
# Feature 3 : Bulk Booking Flag
# ---------------------------------------------------
# Flag bookings with 4 or more tickets.

master_df["Bulk_Booking"] = np.where(
    master_df["Quantity"] >= 4,
    "Yes",
    "No"
)

# ---------------------------------------------------
# Feature 4 : Revenue Per Ticket
# ---------------------------------------------------
# Calculate revenue generated by each individual ticket.

master_df["Revenue_Per_Ticket"] = (
    master_df["Total_Amount"] /
    master_df["Quantity"]
)

# ---------------------------------------------------
# Feature 5 : Seat Price Position
# ---------------------------------------------------

price_ratio = (
    (master_df["Ticket_Price"] - master_df["Min_Price"])
    /
    (master_df["Max_Price"] - master_df["Min_Price"])
)

master_df["Seat_Price_Position"] = np.select(
    [
        price_ratio <= 0.33,
        price_ratio <= 0.66,
        price_ratio > 0.66
    ],
    [
        "Low Range",
        "Mid Range",
        "High Range"
    ],
    default="Unknown"
)

# ---------------------------------------------------
# Feature 6 : Booking Value
# ---------------------------------------------------

master_df["Booking_Value"] = np.select(
    [
        master_df["Total_Amount"] < 2000,
        master_df["Total_Amount"].between(2000, 5000),
        master_df["Total_Amount"] > 5000
    ],
    [
        "Low Value",
        "Medium Value",
        "High Value"
    ],
    default="Unknown"
)
print("="*70)
print("MODULE 4 COMPLETED")
print("="*70)

print("New Features Added:")

new_features = [
    "Booking_Window",
    "Booking_Size",
    "Bulk_Booking",
    "Revenue_Per_Ticket",
    "Seat_Price_Position",
    "Booking_Value"
]

for feature in new_features:
    print("✓", feature)
```

```python
# =============================================================================
# MODULE 5 : REVENUE, PRICING & BUSINESS KPI FEATURES
# Purpose:
# Create executive-level KPIs for revenue analysis, pricing efficiency,
# and business performance reporting.
# =============================================================================

# ---------------------------------------------------
# Feature 1 : Price Utilization Percentage
# ---------------------------------------------------
# Shows how much of the allowed price range was utilized.

master_df["Price_Utilization_Percentage"] = round(
    (master_df["Ticket_Price"] / master_df["Max_Price"]) * 100,
    2
)

# ---------------------------------------------------
# Feature 2 : Revenue Category
# ---------------------------------------------------
# Categorize transactions based on total revenue.

master_df["Revenue_Category"] = np.select(
    [
        master_df["Total_Amount"] < 2000,
        master_df["Total_Amount"].between(2000,5000),
        master_df["Total_Amount"] > 5000
    ],
    [
        "Low Revenue",
        "Medium Revenue",
        "High Revenue"
    ],
    default="Unknown"
)

# ---------------------------------------------------
# Feature 3 : Premium Seat Booking
# ---------------------------------------------------
# Identify Premium and VIP bookings.

master_df["Premium_Seat_Booking"] = np.where(
    master_df["Seat_Category"].isin(["Premium","VIP"]),
    "Yes",
    "No"
)
# ---------------------------------------------------
# Feature 4: Price Range
# ---------------------------------------------------

master_df["Price_Range"] = np.select(
    [
        master_df["Ticket_Price"] <= 1000,
        master_df["Ticket_Price"].between(1001,3000),
        master_df["Ticket_Price"] > 3000
    ],
    [
        "Budget",
        "Mid Range",
        "Premium"
    ],
    default="Unknown"
)

# ---------------------------------------------------
# Feature 5 : Revenue Contribution (%)
# ---------------------------------------------------
# Contribution of each booking to overall revenue.

total_revenue = master_df["Total_Amount"].sum()

master_df["Revenue_Contribution"] = round(
    (master_df["Total_Amount"] / total_revenue) * 100,
    6
)

# ---------------------------------------------------
# Feature 6 : High Value Transaction
# ---------------------------------------------------
# Flag transactions above the 90th percentile.

threshold = master_df["Total_Amount"].quantile(0.90)

master_df["High_Value_Transaction"] = np.where(
    master_df["Total_Amount"] >= threshold,
    "Yes",
    "No"
)

print("="*70)
print("MODULE 5 COMPLETED")
print("="*70)

new_features = [
    "Price_Utilization_Percentage",
    "Revenue_Category",
    "Premium_Seat_Booking",
    "Ticket_Price_Index",
    "Revenue_Contribution",
    "High_Value_Transaction"
]

print("New Features Added:")
for feature in new_features:
    print("✓", feature)
```

```python
# =============================================================================
# MODULE 5 VALIDATION
# =============================================================================

features = [
    "Price_Utilization_Percentage",
    "Revenue_Category",
    "Premium_Seat_Booking",
    "Ticket_Price_Index",
    "Revenue_Contribution",
    "High_Value_Transaction"
]

master_df[features].describe(include="all")
```

```python
# =============================================================================
# MODULE 6 : CUSTOMER SEGMENTATION FEATURES
# Purpose:
# Segment customers based on demographics, income, loyalty,
# and match location to support customer analytics and AI.
# =============================================================================

# ---------------------------------------------------
# Feature 1 : Age Group
# ---------------------------------------------------

master_df["Age_Group"] = pd.cut(
    master_df["Age"],
    bins=[0, 18, 25, 35, 50, 100],
    labels=[
        "Under 18",
        "19-25",
        "26-35",
        "36-50",
        "50+"
    ]
)

# ---------------------------------------------------
# Feature 2 : Income Group
# ---------------------------------------------------

master_df["Income_Group"] = pd.cut(
    master_df["Annual_Income"],
    bins=[0, 500000, 1000000, 2000000, float("inf")],
    labels=[
        "Low Income",
        "Middle Income",
        "Upper Middle",
        "High Income"
    ]
)

# ---------------------------------------------------
# Feature 3 : Loyalty Score
# ---------------------------------------------------

loyalty_mapping = {
    "Bronze": 1,
    "Silver": 2,
    "Gold": 3,
    "Platinum": 4
}

master_df["Loyalty_Score"] = master_df["Loyalty_Tier"].map(loyalty_mapping)

# ---------------------------------------------------
# Feature 4 : Customer Value
# ---------------------------------------------------

master_df["Customer_Value"] = np.select(
    [
        (master_df["Loyalty_Score"] >= 3) &
        (master_df["Total_Amount"] >= 5000),

        (master_df["Loyalty_Score"] >= 2) &
        (master_df["Total_Amount"] >= 2000),

        True
    ],
    [
        "High Value",
        "Medium Value",
        "Low Value"
    ],
    default="Low Value"
)

# ---------------------------------------------------
# Feature 5 : Local Fan
# ---------------------------------------------------

master_df["Local_Fan"] = np.where(
    master_df["City"] == master_df["City_Match"],
    "Yes",
    "No"
)

print("="*70)
print("MODULE 6 COMPLETED")
print("="*70)

new_features = [
    "Age_Group",
    "Income_Group",
    "Loyalty_Score",
    "Customer_Value",
    "Local_Fan"
]

print("New Features Added:")
for feature in new_features:
    print("✓", feature)
```

```python
# =============================================================================
# MODULE 6 VALIDATION
# =============================================================================

features = [
    "Age_Group",
    "Income_Group",
    "Loyalty_Score",
    "Customer_Value",
    "Local_Fan"
]

master_df[features].describe(include="all")
```

```python
# =============================================================================
# MODULE 7 : MATCH INTELLIGENCE & AI FEATURES
# Purpose:
# Create business intelligence features that support executive
# reporting and Gemini AI recommendations.
# =============================================================================

# ---------------------------------------------------
# Feature 1 : Match Stage
# ---------------------------------------------------
# Classify IPL matches into League or Playoffs.

master_df["Match_Stage"] = np.where(
    master_df["Match_Type"] == "League",
    "League",
    "Playoff"
)

# ---------------------------------------------------
# Feature 2 : Seat Category Rank
# ---------------------------------------------------
# Convert seat categories into numeric ranking.

seat_rank = {
    "Economy":1,
    "Premium":2,
    "VIP":3
}

master_df["Seat_Category_Rank"] = master_df["Seat_Category"].map(seat_rank)

# ---------------------------------------------------
# Feature 3 : Booking Urgency
# ---------------------------------------------------
# Categorize bookings based on planning time.

master_df["Booking_Urgency"] = np.select(
    [
        master_df["Booking_Lead_Days"] >= 30,
        master_df["Booking_Lead_Days"].between(15,29),
        master_df["Booking_Lead_Days"] < 15
    ],
    [
        "Planned",
        "Moderate",
        "Urgent"
    ],
    default="Unknown"
)

# ---------------------------------------------------
# Feature 4 : Favorite Team Playing
# ---------------------------------------------------
# Check whether customer's favorite team is playing.

master_df["Favorite_Team_Playing"] = np.where(
    (
        (master_df["Favorite_Team"] == master_df["Home_Team"]) |
        (master_df["Favorite_Team"] == master_df["Away_Team"])
    ),
    "Yes",
    "No"
)

# ---------------------------------------------------
# Feature 5 : Weekend Premium Booking
# ---------------------------------------------------
# Premium/VIP booking on weekends.

master_df["Weekend_Premium_Booking"] = np.where(
    (
        (master_df["Weekend_Match"] == "Yes") &
        (master_df["Seat_Category"].isin(["Premium","VIP"]))
    ),
    "Yes",
    "No"
)

# ---------------------------------------------------
# Feature 6 : Executive Attention
# ---------------------------------------------------
# Flag transactions that deserve management attention.

master_df["Executive_Attention"] = np.where(
    (
        (master_df["High_Value_Transaction"] == "Yes") &
        (master_df["Favorite_Team_Playing"] == "Yes")
    ),
    "High Priority",
    "Normal"
)

print("="*70)
print("MODULE 7 COMPLETED")
print("="*70)

new_features = [
    "Match_Stage",
    "Seat_Category_Rank",
    "Booking_Urgency",
    "Favorite_Team_Playing",
    "Weekend_Premium_Booking",
    "Executive_Attention"
]

print("New Features Added:")
for feature in new_features:
    print("✓", feature)
```

```python
# =============================================================================
# MODULE 7 VALIDATION
# =============================================================================

features = [
    "Match_Stage",
    "Seat_Category_Rank",
    "Booking_Urgency",
    "Favorite_Team_Playing",
    "Weekend_Premium_Booking",
    "Executive_Attention"
]

master_df[features].describe(include="all")
```

```python
# =============================================================================
# MODULE 8 : FINAL VALIDATION & EXPORT
# Purpose:
# Validate the feature engineered dataset and export it
# for Power BI, Gemini AI and Streamlit.
# =============================================================================

print("="*80)
print("FEATURE ENGINEERING SUMMARY")
print("="*80)

print(f"Total Rows              : {master_df.shape[0]}")
print(f"Total Columns           : {master_df.shape[1]}")
print(f"Duplicate Rows          : {master_df.duplicated().sum()}")
print(f"Total Missing Values    : {master_df.isnull().sum().sum()}")

print("\n" + "="*80)
print("DATA TYPES")
print("="*80)
print(master_df.dtypes.value_counts())

print("\n" + "="*80)
print("MISSING VALUES")
print("="*80)

missing = master_df.isnull().sum()

missing = missing[missing > 0]

if len(missing) == 0:
    print("No Missing Values Found ✅")
else:
    print(missing)

print("\n" + "="*80)
print("ENGINEERED FEATURES")
print("="*80)

engineered_features = [

    # Module 3
    "Booking_Lead_Days",
    "Booking_Month",
    "Booking_Day_Name",
    "Match_Month",
    "Match_Day_Name",
    "Match_Quarter",
    "Weekend_Match",

    # Module 4
    "Booking_Window",
    "Booking_Size",
    "Bulk_Booking",
    "Revenue_Per_Ticket",
    "Seat_Price_Position",
    "Booking_Value",

    # Module 5
    "Price_Utilization_Percentage",
    "Revenue_Category",
    "Premium_Seat_Booking",
    "Revenue_Contribution",
    "High_Value_Transaction",
    "Price_Range",

    # Module 6
    "Age_Group",
    "Income_Group",
    "Loyalty_Score",
    "Customer_Value",
    "Local_Fan",

    # Module 7
    "Match_Stage",
    "Seat_Category_Rank",
    "Booking_Urgency",
    "Favorite_Team_Playing",
    "Weekend_Premium_Booking",
    "Executive_Attention"

]

print(f"Total Engineered Features : {len(engineered_features)}\n")

for i, feature in enumerate(engineered_features, start=1):
    print(f"{i:02d}. {feature}")

print("\n" + "="*80)
print("FINAL DATASET SHAPE")
print("="*80)

print(master_df.shape)

print("\nValidation Completed Successfully ✅")
```
