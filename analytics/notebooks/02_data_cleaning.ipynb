# CricIntel AI - Data Cleaning

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)

print("Libraries Imported Successfully")
```

```python
matches = pd.read_csv("../../data/raw/Matches.csv")
customers = pd.read_csv("../../data/raw/Customers.csv")
ticket_sales = pd.read_csv("../../data/raw/Ticket_Sales.csv")

print("Datasets Loaded Successfully")
```

```python
def data_quality_summary(df, dataset_name):
    summary = {
        "Dataset": dataset_name,
        "Rows": df.shape[0],
        "Columns": df.shape[1],
        "Missing Values": df.isnull().sum().sum(),
        "Duplicate Rows": df.duplicated().sum()
    }
    return summary
```

```python
## quality report
quality_report = pd.DataFrame([
    data_quality_summary(matches, "Matches"),
    data_quality_summary(customers, "Customers"),
    data_quality_summary(ticket_sales, "Ticket Sales")
])

display(quality_report)
```

```python
## missing values
print("Missing Values - Matches")
display(matches.isnull().sum())
```

```python
print("Missing Values - Ticket Sales")
display(ticket_sales.isnull().sum())
```

```python
## Check Duplicate Records matchess
print("Duplicate Records - Matches")

duplicate_matches = matches[matches.duplicated()]

print(f"Total Duplicate Records: {duplicate_matches.shape[0]}")

display(duplicate_matches)
```

```python
## Check duplicate records customer
print("Duplicate Records - Customers")

duplicate_customers = customers[customers.duplicated()]

print(f"Total Duplicate Records: {duplicate_customers.shape[0]}")

display(duplicate_customers)
```

```python
# Check dupllicate records for ticket sales 
print("Duplicate Records - Ticket Sales")

duplicate_ticket_sales = ticket_sales[ticket_sales.duplicated()]

print(f"Total Duplicate Records: {duplicate_ticket_sales.shape[0]}")

display(duplicate_ticket_sales)
```

```python
## Validating match ids
print("Checking Unique Match IDs")

print("Total Match IDs :", matches["Match_ID"].count())
print("Unique Match IDs:", matches["Match_ID"].nunique())
```

```python
## valiprint("Checking Unique Customer IDs")
print("Checking Unique Customer IDs")

print("Total Customer IDs :", customers["Customer_ID"].count())
print("Unique Customer IDs:", customers["Customer_ID"].nunique())
```

```python
## Validate Ticket IDs
print("Checking Unique Ticket IDs")

print("Total Ticket IDs :", ticket_sales["Ticket_ID"].count())
print("Unique Ticket IDs:", ticket_sales["Ticket_ID"].nunique())
```

```python
## Validating datatypes
print("Matches Data Types")
display(matches.dtypes)

print("\nCustomers Data Types")
display(customers.dtypes)

print("\nTicket Sales Data Types")
display(ticket_sales.dtypes)
```

```python
# Create working copies for data cleaning

matches_clean = matches.copy()
customers_clean = customers.copy()
ticket_sales_clean = ticket_sales.copy()

print("Working copies created successfully.")
```

```python
# Convert date columns

matches_clean["Match_Date"] = pd.to_datetime(matches_clean["Match_Date"])

customers_clean["Registration_Date"] = pd.to_datetime(
    customers_clean["Registration_Date"]
)

ticket_sales_clean["Booking_Date"] = pd.to_datetime(
    ticket_sales_clean["Booking_Date"]
)

print("Date conversion completed.")
```

```python
print(matches_clean["Match_Date"].dtype)

print(customers_clean["Registration_Date"].dtype)

print(ticket_sales_clean["Booking_Date"].dtype)
```

```python
print("Gender Values")

display(customers_clean["Gender"].value_counts())
```

```python
print("Loyalty Tier")

display(customers_clean["Loyalty_Tier"].value_counts())
```

```python
print("Booking Status")

display(ticket_sales_clean["Booking_Status"].value_counts())
```

```python
print("Payment Methods")

display(ticket_sales_clean["Payment_Method"].value_counts())
```

```python
print("Seat Categories")

display(ticket_sales_clean["Seat_Category"].value_counts())
```

```python
invalid_quantity = ticket_sales_clean[
    ticket_sales_clean["Quantity"] <= 0
]

print("Invalid Ticket Quantity :", len(invalid_quantity))

display(invalid_quantity)
```

```python
invalid_price = ticket_sales_clean[
    ticket_sales_clean["Ticket_Price"] <= 0
]

print("Invalid Ticket Price :", len(invalid_price))

display(invalid_price)
```

```python
invalid_amount = ticket_sales_clean[
    ticket_sales_clean["Total_Amount"] <= 0
]

print("Invalid Total Amount :", len(invalid_amount))

display(invalid_amount)
```

```python
# Validate Customer IDs

invalid_customers = ticket_sales_clean[
    ~ticket_sales_clean["Customer_ID"].isin(customers_clean["Customer_ID"])
]

print("Invalid Customer IDs :", len(invalid_customers))

display(invalid_customers)
```

```python
# Validate Match IDs

invalid_matches = ticket_sales_clean[
    ~ticket_sales_clean["Match_ID"].isin(matches_clean["Match_ID"])
]

print("Invalid Match IDs :", len(invalid_matches))

display(invalid_matches)
```

```python
# Clean email addresses

customers_clean["Email"] = (
    customers_clean["Email"]
    .str.strip()
    .str.replace("’", "", regex=False)
    .str.replace("'", "", regex=False)
)

print("Email cleaning completed.")
```

```python
# Validate Email Format

invalid_email = customers_clean[
    ~customers_clean["Email"].str.contains(
        r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$",
        regex=True
    )
]

print("Invalid Email Addresses :", len(invalid_email))

display(invalid_email)
```

```python
# Validate Phone Numbers

invalid_phone = customers_clean[
    customers_clean["Phone"].astype(str).str.len() != 10
]

print("Invalid Phone Numbers :", len(invalid_phone))

display(invalid_phone)
```

```python
# Validate Registration Dates

future_registration = customers_clean[
    customers_clean["Registration_Date"] > pd.Timestamp.today()
]

print("Future Registration Dates :", len(future_registration))

display(future_registration)
```

```python
# Validate Match Year

invalid_match_year = matches_clean[
    matches_clean["Match_Date"].dt.year != 2026
]

print("Invalid Match Year :", len(invalid_match_year))

display(invalid_match_year)
```

```python
# Validate Stadium Capacity

capacity_check = matches_clean[
    (
        matches_clean["VIP_Capacity"]
        + matches_clean["Premium_Capacity"]
        + matches_clean["Economy_Capacity"]
    )
    != matches_clean["Stadium_Capacity"]
]

print("Capacity Mismatch :", len(capacity_check))

display(capacity_check)
```

```python
# Final Data Quality Report

quality_summary = pd.DataFrame({
    "Validation": [
        "Missing Values",
        "Duplicate Rows",
        "Invalid Customer IDs",
        "Invalid Match IDs",
        "Invalid Emails",
        "Invalid Phone Numbers",
        "Future Registration Dates",
        "Invalid Match Year",
        "Capacity Mismatch"
    ],
    "Issues Found": [
        matches_clean.isnull().sum().sum()
        + customers_clean.isnull().sum().sum()
        + ticket_sales_clean.isnull().sum().sum(),

        matches_clean.duplicated().sum()
        + customers_clean.duplicated().sum()
        + ticket_sales_clean.duplicated().sum(),

        len(invalid_customers),
        len(invalid_matches),
        len(invalid_email),
        len(invalid_phone),
        len(future_registration),
        len(invalid_match_year),
        len(capacity_check)
    ]
})

display(quality_summary)
```

```python
# Overall Validation Status

if quality_summary["Issues Found"].sum() == 0:
    print("All data quality checks passed successfully.")
else:
    print("Data quality issues detected. Review the validation report.")
```
