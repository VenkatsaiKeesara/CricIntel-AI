```python
import pandas as pd
import numpy as np

# Display all columns
pd.set_option("display.max_columns", None)

# Display wider tables
pd.set_option("display.width", 1000)

print("Libraries Imported Successfully")
```

```python
# Load datasets

matches = pd.read_csv("../../data/raw/Matches.csv")
customers = pd.read_csv("../../data/raw/Customers.csv")
ticket_sales = pd.read_csv("../../data/raw/Ticket_Sales.csv")

print("Datasets Loaded Successfully")
```

```python
print("Matches Dataset Shape :", matches.shape)
print("Customers Dataset Shape :", customers.shape)
print("Ticket Sales Dataset Shape :", ticket_sales.shape)
```

```python
print("Matches Dataset")
display(matches.tail())

print("Customers Dataset")
display(customers.tail())

print("Ticket Sales Dataset")
display(ticket_sales.tail())
```
